```python
import matplotlib.pyplot as plt
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"], errors="coerce")

```

```python
monthly_sales = df.groupby(df["Order Date"].dt.month)["Sales"].sum()
```

```python
monthly_sales
```

```python
monthly_sales.plot()
```

```python
daily_Profit = df.groupby(df["Order Date"].dt.day)["Profit"].sum()
```

```python
daily_Profit.plot()
```

```python
yearly_profit = df.groupby(df["Order Date"].dt.)["Sales"].sum()
```

```python
yearly_profit.plot()
```

```python
yearly_profit.plot(kind="bar")
```

```python
yearly_profit.plot(kind="pie")
```
